```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
target_URL = 'https://www.ncaa.com/stats/volleyball-men/d1/current/individual/1124/p2'
web_page = requests.get(target_URL)
html_code = web_page.content
parsed_code = BeautifulSoup(html_code,"html.parser")
```

```python
table_body=parsed_code.tbody
```

```python
player_rows=table_body.find_all('tr')
```

```python
len(player_rows)
```

```python
td_tags=player_rows[0].find_all('td')
```

```python
td_tags
```

```python
player_info = []
for info in td_tags:
    player_info.append(info.get_text())
print(player_info)
```

```python
th_tags=parsed_code.find_all('th')
th_tags
```

```python
col_names=[]
for col in th_tags:
    col_names.append(col.get_text())
print(col_names)
```

```python
Data=pd.DataFrame(columns=col_names)
Data
```

```python
for player in player_rows:
    td_tags=player.find_all('td')
    player_info=[]
    for info in td_tags:
        player_info.append(info.get_text())
    Data.loc[len(Data)] = player_info
```

```python
Data
```

```python
Data.to_excel('Mens Attack Volleyball2.xlsx', index='false')
```
